```python
import os
import shutil

# Source and destination directories
src_dir = 'C:/Users/yalmo/PycharmProjects/pythonProject1/data/images'
dst_dir = 'C:/Users/yalmo/PycharmProjects/pythonProject1/data/labels'

# Get a list of all files in the source directory
files = os.listdir(src_dir)
print(files)
```

```python


# Iterate over the files and move JSON files to the destination directory
for file_name in files:
    print(file_name)
    src_file = os.path.join(src_dir, file_name)
    print(src_file)
    if file_name.endswith('.json'):
        dst_file = os.path.join(dst_dir, file_name)
        print(dst_file)
        shutil.move(src_file, dst_file)
```

```python
import os

# replace 'path/to/parent/directory' with the path to the parent directory where you want to create the train/val/test directories
parent_directory = 'C:/Users/yalmo/PycharmProjects/pythonProject1/data'

# create the train/val/test directories
os.makedirs(os.path.join(parent_directory, 'train', 'images'))
os.makedirs(os.path.join(parent_directory, 'train', 'labels'))
os.makedirs(os.path.join(parent_directory, 'val', 'images'))
os.makedirs(os.path.join(parent_directory, 'val', 'labels'))
os.makedirs(os.path.join(parent_directory, 'test', 'images'))
os.makedirs(os.path.join(parent_directory, 'test', 'labels'))

# confirm that directories were created
for dir_name in ['train', 'val', 'test']:
    for subdir_name in ['images', 'labels']:
        subdirectory = os.path.join(parent_directory, dir_name, subdir_name)
        if os.path.exists(subdirectory):
            print(f"Directory '{subdirectory}' created successfully!")
        else:
            print(f"Directory '{subdirectory}' was not created.")
```

```python
import os
import random
import shutil

# Set the path to the data directory
data_dir = 'C:/Users/yalmo/PycharmProjects/pythonProject1/data'

# Set the number of files to select
num_files = 50

# Set the path to the images and labels directories
images_dir = os.path.join(data_dir, 'images')
labels_dir = os.path.join(data_dir, 'labels')

# Set the path to the train directory
train_dir = os.path.join(data_dir, 'train')

# Create the train/images and train/labels directories if they don't exist
train_images_dir = os.path.join(train_dir, 'images')
train_labels_dir = os.path.join(train_dir, 'labels')
os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)

# Get a list of all the JSON files in the labels directory
json_files = [f for f in os.listdir(labels_dir) if f.endswith('.json')]

# Randomly select the specified number of files
selected_files = random.sample(json_files, num_files)

# Copy the selected JSON files to the train/labels directory
for file in selected_files:
    src_path = os.path.join(labels_dir, file)
    dst_path = os.path.join(train_labels_dir, file)
    shutil.copy(src_path, dst_path)

    # Copy the corresponding PNG file to the train/images directory
    png_file = file.replace('.json', '.png')
    src_path = os.path.join(images_dir, png_file)
    dst_path = os.path.join(train_images_dir, png_file)
    shutil.copy(src_path, dst_path)
```
